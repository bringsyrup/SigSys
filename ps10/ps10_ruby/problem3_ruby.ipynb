### B). $\frac{s}{s^2 + 100s + 1}$

```python
combinedplot(signal.lti([1, 0],[1, 100, 1]),T=np.arange(0,200,.1),w=np.logspace(-2,2,100),axis='tight')
```

This a passband filter b/c we have two poles, one near -100 and a pole slightly less than 0. This means the step response will forst quickly try to converge to 0.01 due to the large magnitude pole (around Re{s} = -100), but is then affected by the pole near zero and slowly converges to a value less than 0.01. 

### C). $\frac{s}{s^2 + s + 1}$

```python
combinedplot(signal.lti([1, 0],[1, 1, 1]),T=np.arange(0,200,.1),w=np.logspace(-2,2,100),axis='tight')
```

This is another band-pass filter b/c there are two poles. The step response oscillates b/c the poles are imaginary, but still converges b/c the real component is less than 0! 

### D). $\frac{s}{s^2 + 0.1s + 1}$

```python
combinedplot(signal.lti([1, 0],[1, .1, 1]),T=np.arange(0,200,.1),w=np.logspace(-2,2,100),axis='tight')
```

Alright, another BPF cuz there are 2 poles. This time we get more oscillation b/c the real components of the poles are closer to 0, and we still get convergence b/c the real components are less than zero. The frequency response has a narrower band-gap too due to the imaginary components being farther apart(?).

### E). $\frac{s^2 - 0.01s + 1}{s^2 + 0.01s + 1}$

```python
combinedplot(signal.lti([1, -0.01, 1],[1, .01, 1]),T=np.arange(0,200,.1),w=np.logspace(-2,2,100),axis='tight')
```

This kind of looks like a BSF if you just look at the pole-zero and step response graphs because not only are there 2 poles, there are also 2 zeros! BUT the bode plot shows no stop band because the holes are on the opposite side of the imaginary axis (relative to the poles) in the pole-zero graph. So it's just an all-pass filter! We also have convergence on the step reponse b/c the poles are on the left.
